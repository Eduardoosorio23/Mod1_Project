```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
genre_percent = pd.read_csv('data/genre_percent_per_year.csv')
```

```python
top_votes_per_genre = pd.read_csv('data/top_votes_per_genre_tmdb.csv', index_col=[0])
```

```python
popularity_per_genre = pd.read_csv('data/popularity_per_genre_tmdb.csv')
```

```python
enthusiasm = pd.read_csv('data/enthusiasm_imdb.csv')
```

```python
top_votes_per_genre.set_index('name', inplace=True)
```

```python
top_votes_per_genre.sort_values(by=['vote_average'], inplace=True)
```

```python
sns.set(style="darkgrid")
sns.set_palette("husl", 60)
fig_dims = (20, 10)
fig, ax = plt.subplots(figsize=fig_dims)
ax.set(ylim=(0, 8))
ax = sns.barplot(x=top_votes_per_genre.index, y=top_votes_per_genre.vote_average, ax=ax, data=top_votes_per_genre)
ax.set(xlabel="GENRE", ylabel="VOTE AVERAGE")
ax.set_title("Vote Averages per Genre", fontsize = 20)
plt.savefig('images/vote_averages_per_genre.png')
plt.show()
```

```python
enthusiasm.sort_values(by=['enthusiasm_score'], inplace=True)
```

```python
sns.set(style="darkgrid")
fig_dims = (20, 10)
fig, ax = plt.subplots(figsize=fig_dims)
ax.set(ylim=(0, 33))
ax = sns.barplot(x=enthusiasm.genre, y=enthusiasm.enthusiasm_score, ax=ax, data=enthusiasm)
ax.set(xlabel="GENRE", ylabel="ENTHUSIASM SCORE")
ax.set_title("Enthusiasm per Genre", fontsize = 20)
plt.savefig('images/enthusiasm_per_genre.png')
plt.show()
```

```python
popularity_per_genre.sort_values(by=['popularity'], inplace=True)
```

```python
sns.set(style="darkgrid")
sns.set_palette("husl", 50)
fig_dims = (20, 10)
fig, ax = plt.subplots(figsize=fig_dims)
ax.set(ylim=(0, 22))
ax = sns.barplot(x=popularity_per_genre.name, y=popularity_per_genre.popularity, ax=ax, data=popularity_per_genre)
ax.set(xlabel="GENRE", ylabel="POPULARITY SCORE")
ax.set_title("Popularity per Genre", fontsize = 20)
plt.savefig('images/popularity_per_genre.png')
plt.show()
```
