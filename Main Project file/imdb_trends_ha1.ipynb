```python
# open df of movie by their regions
regions = pd.read_csv('data/title.akas.csv')
regions
```

```python
# limit movies to US
us_movies = regions.loc[(regions.region == 'US')]
```

```python
# merge the movies and ratings
rated_movies = pd.merge(movies, ratings)
```

```python
rated_movies.head()
```

```python
rated_movies.shape
```

```python
# merge that df with the US movies
us_movies = rated_movies.merge(us_movies, left_on="primary_title", right_on="title")
```

```python
us_movies.head()
```

```python
us_movies = us_movies.filter(['title', 'genres', 'numvotes', 'start_year'])
```

```python
us_movies = us_movies.drop_duplicates(subset=["title", "genres", "numvotes", "start_year"])
```

```python
us_movies.sort_values(by=["numvotes"], ascending = False).head()
```

```python
# drop movies that are little seen (fewer than 100 votes)
us_movies = us_movies.loc[us_movies.numvotes > 100]
```

```python
# drop NaN values
us_movies.dropna(inplace = True)
```

```python
# function to divide up the genres 
def split_genres(string):
    genres = list(string.split(','))
    new_list = []
    for genre in genres:
        new_list.append(genre)
    return new_list
```

```python
us_movies.head()
```

```python
# create df with one genre per row

genre_year = pd.DataFrame()
columns = ['title', 'genre', 'numvotes', 'start_year']
for i in range(len(us_movies)):
    genres = split_genres(us_movies.iloc[i]['genres'])
    for j in range(len(genres)):
        title = us_movies.iloc[i]['title']
        numvotes = us_movies.iloc[i]['numvotes']
        genre = genres[j]
        start_year = us_movies.iloc[i]['start_year']
        values = [title, genre, numvotes, start_year]
        zipped = zip(columns, values)
        to_append = dict(zipped)
        genre_year = genre_year.append(to_append, True)
genre_year.head()
```

```python
# renaming the year column
genre_year = genre_year.rename(columns={'start_year': 'year'})
```

```python
# making the year column an integer
genre_year['year'] = genre_year['year'].astype(int)
```

```python
# seeing how many movies in total are made each year
genre_year.year.value_counts()
```

```python
# again, seeing how many movies in total are made each year
genre_year.groupby('year').count()['genre']
```

```python
# making a dictionary of how many movies are made each year
movies_per_year = genre_year.groupby('year').count()['genre'].to_dict()
```

```python
# making a dictionary of how many of each genre is made each year
genre_per_year = genre_year.groupby(['year', 'genre']).count().to_dict()
```

```python
# the dictionary was nested under 'numvotes'. Removing that. 
genre_per_year = genre_per_year.get('numvotes')
```

```python
# testing getting a single value from a genre and year. Here 161 action movies were made in 2010
genre_per_year.get((2010, 'Action'))
```

```python
# finding the length of the dictionary. It's 220
len(genre_per_year)
```

```python
# converting the diciontary into a sorted list
genre_per_year = sorted(genre_per_year.items())
```

```python
genre_per_year[0]
```

```python
# changing the list to putting genre first in tuple and getting rid of useless genres
list = []
for x in range(len(genre_per_year)):
    genre = genre_per_year[x][0][1]
    year = genre_per_year[x][0][0]
    percent = genre_per_year[x][1]
    if (genre != 'Adult' and genre != "News" and genre != 'Reality-TV' and genre != "Game-Show"):
        list.append([genre, (year , percent)])
genre_per_year = list
```

```python
# sorting the list alphabetically
genre_per_year = sorted(genre_per_year)
```

```python
# checking how many years are in each genre and that they're equal
counter = {}
for x in range(len(genre_per_year)):
    if genre_per_year[x][0] in counter:
        counter[genre_per_year[x][0]] += 1
    else:
        counter[genre_per_year[x][0]] = 1
counter
```

```python
# Defining a function to get the percentage of how many movies were made that year
def percent_genre_made_per_year(genre_year_number):
    year = genre_year_number[1][0]
    number = genre_year_number[1][1]
    genre_movies_made_that_year = number
    total_num_movies_made_that_year = movies_per_year.get(year)
    return round((genre_movies_made_that_year/total_num_movies_made_that_year * 100), 1)
```

```python
percent_genre_made_per_year(genre_per_year[0])
```

```python
# test out a for loop for each genre-year
list = []
for i in range(10):
    percent = percent_genre_made_per_year(genre_per_year[i])
    list.append(percent)
list
```

```python
# function to help add ten and cycle through the list
def add_ten(numbers):
    x = numbers[0]
    y = numbers[1]
    x += 10
    y += 10
    return x, y
```

```python
# test code to cycle through the genre_per_year list and pull out the percentages
other_list = []
start_stop = (0, 10)
while start_stop[1] <= len(genre_per_year):
    temp_list = []
    for i in range(start_stop[0], start_stop[1]):
        percent = percent_genre_made_per_year(genre_per_year[i])
        temp_list.append(percent)
    other_list.append(temp_list)
    start_stop = add_ten(start_stop)
len(other_list)
```

```python
# populating column names with genre name and years
columns = ['genre']
for x in range(2010, 2020):
    columns.append(str(x))
```

```python
# intializing new dataframe
genre_percent = pd.DataFrame()

# incrementing by ten because each genre has 10 years of data 
start_stop = (0, 10)
while start_stop[1] <= len(genre_per_year):
    temp_list = []
    for i in range(start_stop[0], start_stop[1]):
        percent = percent_genre_made_per_year(genre_per_year[i])
        temp_list.append(percent)
        genre_name = genre_per_year[i][0]
    temp_list.insert(0, genre_name)
    values = temp_list
    zipped = zip(columns, values)
    to_append = dict(zipped)
    genre_percent = genre_percent.append(to_append, True)
    start_stop = add_ten(start_stop)
genre_percent = genre_percent.set_index('genre')
genre_percent
```
