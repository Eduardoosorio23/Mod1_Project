```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
profitability = pd.read_csv('profitability.csv')#uploaded data
profitability
```

```python
profitability.drop(labels='Unnamed: 0',axis=1, inplace = True)# dropped unnamed column
profitability
```

```python
 pd.set_option('float_format', '{:f}'.format)
# I added this line in to show entire numbers instead of numbers ^x. 
#It seemed to adjusted it the everything even before this line.
```

```python
top_domestic_gross = profitability.sort_values(by="domestic_gross", ascending=False).reset_index(drop=True).head(25)
top_domestic_gross# To find out what the top 25 domestic gross movies are
```

```python
top_foreign_gross = profitability.sort_values(by="foreign_gross", ascending=False).reset_index(drop=True).head(25)
top_foreign_gross# to find out what the top 25 foriegn movies are
```

```python
studio_gross_domestic = profitability.groupby('studio').sum().sort_values(by='domestic_gross', ascending=False).filter(['domestic_gross', 'production_budget', 'profitability']).reset_index()
studio_gross_domestic #top earning studios in the domestic market
```

```python
studio_gross_foreign  =  profitability.groupby('studio').sum().sort_values(by='foreign_gross', ascending=False).filter(['foreign_gross', 'production_budget', 'profitability']).reset_index()
studio_gross_foreign #Top earning studio in the foreign market
```

```python
most_profitable_studio = profitability.groupby('studio').sum().sort_values(by='profitability', ascending=False).filter(['domestic_gross', 'foreign_gross', 'production_budget', 'profitability']).reset_index()
most_profitable_studio
```

```python
top_ten_most_profitable_studio = profitability.groupby('studio').sum().sort_values(by='profitability', ascending=False).filter(['domestic_gross', 'foreign_gross', 'production_budget', 'profitability']).reset_index()
top_ten_most_profitable_studio
```

```python
gf = most_profitable_studio.melt(id_vars=['studio'], value_vars=['domestic_gross', 'foreign_gross'], value_name='Gross Dollar Amount')
gf
```

```python
sns.catplot(x="studio", y="Gross Dollar Amount", hue="variable", data=gf,
                height=6, kind="bar", palette="muted")
```
