```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
profitability_total = pd.read_csv('profitability_total.csv')#uploaded data
profitability_total
```

```python
titles = pd.read_csv('tmbd_titles.csv')
titles
```

```python
genre_profit = pd.merge(profitability_total,
                   titles[['genre_ids','title']],
                   on='title', how='left')
genre_profit# merged my profitability_total DF to my titles DF
```

```python
genre_profit = genre_profit.drop_duplicates(subset=['title'], inplace=False)
```

```python
 pd.set_option('float_format', '{:f}'.format)
# I added this line in to show entire numbers instead of numbers ^x. 
#It seemed to adjusted it the everything even before this line.
```

```python
genre_profit
```

```python
genre_profit.groupby("genre_ids").sum().sort_values(by="profitability", ascending=False)
#sorted by Genre Ids but they were embeded in lists inside the column
```

```python
def clean_column(x):# I'm going to use this function to clean out the string
  
    if isinstance(x, str):
        return(x.replace('[', '').replace(']', '').replace(',', ''))
    return(x)#modified one of my previous functions to clean the strings in genre_ids
```

```python
genre_profit['genre_ids'] = genre_profit['genre_ids'].apply(clean_column)
```

```python
genre_profit.head()#strings are cleaned
```

```python
genre_id = pd.DataFrame(genre_profit.genre_ids.str.split(' '))#split the strings in genre_ids
```

```python
genre_id.genre_ids[0]#stings are split
```

```python
genre_profit.genre_ids = genre_id.genre_ids
```

```python
genre_profit
```

```python
genre_profit = genre_profit.genre_ids.apply(pd.Series)\
.merge(genre_profit, left_index = True, right_index = True)\
.drop(["genre_ids"], axis = 1) \
.melt(id_vars = ['title', 'profitability'], value_vars= [0,1,2,3,4,5,6], value_name = "genre_ids")\
.sort_values(by='title')
# made a new series of just the genre ids
# merged the old table on to the new genre ids table
# dropped the old "genre_ids" column
# used the melt function to create a new table that categorizes each movie to its repective genres
# sorted them by title
```

```python
genre_profit
```

```python
genre_profit.rename(columns =  {'genre_ids': 'id'}, inplace = True)# renamed genre_ids to id
```

```python
genre_profit = genre_profit.dropna(axis=0, how='all')# dropped the null values
```

```python
genre_profit = genre_profit.groupby(['id']).sum().reset_index()
genre_profit# for some reason genre_profit.id[0] wouldnt drop with the dropna function
```

```python
genre_profit = genre_profit.drop([0])
genre_profit# manuallly dropped the row
```

```python
genre_profit.info()
```

```python
genre_profit['id'] = genre_profit['id'].astype('float')# now that there are no NaN values I can convert to a float
```

```python
genre_codes = pd.read_csv('genre_codes.csv')
genre_codes #uploaded the genre codes table
```

```python
genre_codes = genre_codes.drop(columns='Unnamed: 0')
genre_codes # Dropped  unnamed column
```

```python
genre_codes.id = genre_codes.id.astype('float') # changed id to float
```

```python
genre_codes.info()
```

```python
most_profitable_genre = pd.merge(genre_codes,
                   genre_profit,
                   on='id', how='outer')
most_profitable_genre #merged genre_codes to my genre_profit table
```

```python
most_profitable_genre = most_profitable_genre.drop(labels='id', axis=1) # didnt need the id column anymore 
```

```python
most_profitable_genre.info()
```

```python
most_profitable_genre
```

```python
most_profitable_genre = most_profitable_genre.rename(columns =  {'name': 'genre'}, inplace = False) # renamed the name column to genre
```

```python
most_profitable_genre
```

```python
most_profitable_genre = most_profitable_genre.groupby('genre').sum().sort_values(by='profitability', ascending=False).reset_index()
most_profitable_genre # sorted the values by profitability
```

```python
top_5_genre = most_profitable_genre.head(5) # grabbed the top five genres
```

```python

ax = sns.catplot(x="genre", y="profitability", data=top_5_genre,
                height=8, aspect=1, kind="bar", palette="muted")
```
