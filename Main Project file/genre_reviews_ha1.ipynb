```python
# creating function to help see range of vote_counts
def divide_list(list, num1, num2):
    new_list = []
    for x in list:
        if (x > num1) & (x < num2):
            new_list.append(x)
    return new_list
```

```python
# creating new DF for genre popularity

genre_pop = pd.DataFrame(columns=['popularity', 'title', 'vote_average', 'genre'])
```

```python
genre_pop
```

```python
# looking at the first movie 

movies.iloc[0:1]
```

```python
# genre_ids are list of numbers, actually in a string. 
movies.iloc[0]['genre_ids']
```

```python
# function to isolate numbers

def split_ids(string):
    string = string[1:-1]
    numbers = list(string.split(','))
    new_list = []
    for i in numbers:
        new_list.append(int(i))
    return new_list
    
```

```python
# testing function
split_ids(movies.iloc[0]['genre_ids'])
```

```python
# populating new DF using nested for loop
columns = ['title', 'genre', 'vote_average', 'popularity']
for i in range(len(movies)):
    ids = split_ids(movies.iloc[i]['genre_ids'])
    for j in range(len(ids)):
        popularity = movies.iloc[i]['popularity']
        title = movies.iloc[i]['title']
        avg = movies.iloc[i]['vote_average']
        genre = int(ids[j])
        values = [title, genre, avg, popularity]
        zipped = zip(columns, values)
        to_append = dict(zipped)
        genre_pop = genre_pop.append(to_append, True)
    
genre_pop
```

```python
# here are all the different genre codes
genre_pop.genre.unique()
```

```python
# dropping all rows with NaN in genre
genre_pop.dropna(subset= ["genre"], inplace = True)
```

```python
genre_pop.shape
```

```python
genre_pop.head()
```

```python
# new DF to rank movies with highest mean vote_average
top_votes = genre_pop.groupby("genre").mean().sort_values(by="vote_average", ascending=False).reset_index()
top_votes
```

```python
# new DF to rank movies with highest mean popularity
top_popularity = genre_pop.groupby("genre").mean().sort_values(by="popularity", ascending=False).reset_index()
top_popularity
```
